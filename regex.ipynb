```python
import re
txt="Hello world"
```

```python
x=re.findall("[a-m]",txt)
print(x)
```

.findall() is used to find atleast one matching character from the provided list
here in this code it has list of a-m that means a to m lowercase characters should be there

```python
#search for a sequence starts with He than 2 random characters and ends with o
x=re.findall("He..o",txt)
print(x)
```

.. means it contain all the characters also special characters except \n(new line)

```python
#check if string starts with hello
x=re.findall("^Hello",txt)
if x:
  print("yes, string starts with 'hello'")
else:
  print("no match")
```

```python
#check if string ends with world
x=re.findall("world$",txt)
if x:
  print("yes, string ends with world")
else:
  print("no match")
```

```python
#search for a sequence that starts with he, followed by any character and ends with r
x=re.findall("He.*r",txt)
print(x)
```

```python
#search for a sequence that starts with he, followed by 1 or more any character and ends with o
x=re.findall("He.+o",txt)
print(x)
```

```python
#search for a sequence that starts with he, followed by zero or one any character and ends with o
x=re.findall("He.?o",txt)        #2 character between he and o
print(x)
```

```python
#search for a sequence that starts with he, followed by any exact two character and ends with o
x=re.findall("He.{2}o",txt)
print(x)
```

```python
txt="the rain in spain falls mainly in the plain!"
#check if the strings contains either falls or stays
x=re.findall("falls|stays",txt)
print(x)
```

```python
import re
txt="Åland"

#find all ascii matches
print(re.findall("\w",txt,re.ASCII))

#without the flag, the example would return all character
print(re.findall("\w",txt))

#same result using shorthand re.A flag
print(re.findall("\w",txt,re.A))
```

```python
txt="rain in spain"
print(re.findall("spain",txt,re.DEBUG))
```

```python
txt="""hi
my
name
is
rizz"""

#search for sequence that starts with 'me', followed by one random or even newline character and continue with "is":
print(re.findall("me.is",txt,re.DOTALL))

#this example would return no match without the re.dotall flag
print(re.findall("me.is",txt))

#same result with the shorthand re.S flag
print(re.findall("me.is",txt,re.S))
```

```python
txt="""there
aint much
rain in
spain"""

#search for the sequence "ain" at the beginning of the line:
print(re.findall("^ain",txt,re.MULTILINE))

#without the flag, the example would return no match
print(re.findall("^ain",txt))

#same result using the shorthand re.M
print(re.findall("^ain",txt,re.M))

```

```python
txt="the rain in spain falls mainly on the plain"

#find and return words that contains the phrase "ain"

pattern="""
[A-Za-z]* #starts with any letter
ain+      #contains 'ain'
[a-z*]    #FOLLOWED BY ANY SMALL LETTER
"""

print(re.findall(pattern,txt,re.VERBOSE))

#the ex would return nothing without the re.verbose flag
print(re.findall(pattern,txt))

#same result with the shorthand re.X flag
print(re.findall(pattern,txt,re.X))
```
